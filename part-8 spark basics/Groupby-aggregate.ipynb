```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('aggs').getOrCreate()
```

```python
df = spark.read.csv('sales_info.csv', inferSchema=True,header=True)
```

```python
df.show()
```

```python
df.printSchema()
```

```python
df.groupBy("Company")
```

```python
df.groupBy("Company").mean().show()
df.groupBy("Company").sum().show()
df.groupBy("Company").count().show()
```

```python
df.agg({'Sales':'sum'}).show()
```

```python
df.agg({'Sales':'max'}).show()
```

```python
group_data = df.groupBy("Company").agg({'Sales':'max'}).show()
#group_data.agg({'Sales':'max'}).show()
group_data
```

```python
from pyspark.sql.functions import countDistinct, avg,stddev
```

```python
df.select(countDistinct('Sales')).show()
```

```python
df.select(avg('Sales')).show()
```

```python
df.select(avg('Sales').alias('Average sales')).show()
```

```python
df.select(stddev('Sales').alias('Standard Deviation')).show()
```

```python
from pyspark.sql.functions import format_number
```

```python
sales_std = df.select(stddev('Sales').alias('Standard Deviation'))
sales_std.select(format_number('Standard Deviation',2).alias('final STD')).show()
```
